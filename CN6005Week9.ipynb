```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = [
    ['Mango', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Doll', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo'],
    ['Mango', 'Apple', 'Key-chain', 'Eggs'],
    ['Mango', 'Umbrella', 'Corn', 'Key-chain', 'Yo-yo'],
    ['Corn', 'Onion', 'Onion', 'Key-chain', 'Ice-cream', 'Eggs']
]
```

```python
transactions = [list(set(t)) for t in transactions]
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_array, columns=te.columns_)
print("One-hot encoded data:")
print(df_encoded)
```

```python
# min_support = 60% = 0.6
frequent_itemsets = apriori(df_encoded,
                            min_support=0.6,
                            use_colnames=True)

# Sort just for nicer display
frequent_itemsets = frequent_itemsets.sort_values(by=['support', 'itemsets'],
                                                  ascending=[False, True])

print("\nFrequent itemsets (support >= 0.6):")
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.8)

# Keep only rules with at least 1 item in antecedent and consequent (usually true anyway)
rules = rules[
    (rules['antecedents'].apply(lambda x: len(x) >= 1)) &
    (rules['consequents'].apply(lambda x: len(x) >= 1))
]

# Select and sort main columns
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules = rules.sort_values(by=['confidence', 'lift', 'support'],
                          ascending=[False, False, False])

print("\nStrong association rules (conf >= 0.8):")
print(rules)
```

```python
import matplotlib.pyplot as plt

# Count how many transactions contain each item
item_counts = df_encoded.sum(axis=0).sort_values(ascending=False)

# Top 10 (or fewer, since you have few items)
top_items = item_counts.head(10)

plt.figure(figsize=(8, 4))
top_items.plot(kind='bar', title='Top 10 Most Purchased Items')
plt.xlabel("Item")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
